# Converting location files

## Spain

```python
header_list = ["postcodes", "poblacionid", "town", "latitude", "longitude", "provinciaid", "region"]
```

```python
df = pd.read_csv("../../../../Downloads/postal.cat.csv", sep=";", names=header_list, header=0)
```

```python
df.head()
```

```python
df_final = df[["postcodes", "town", "latitude", "longitude", "region"]]
```

```python
df_final['country_iso']='esp'
```

```python
df_final['uk_region']=None
```

```python
list(df_final.columns)
```

```python
df_final.to_csv('location_files/spain.csv',  index = False)
```
